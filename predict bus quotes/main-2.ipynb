```python
import numpy as np
import pandas as pd
import scipy.stats as ss
import seaborn as sns
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
df =  pd.read_csv("./final_dataset.csv",index_col=0)
df
```

```python
features = df.drop(labels=['Supplier Quote(including GST)','Number of MiniBus','Number of Medium Coach','Number of Large MiniBus','Number of Small Coach','Extra Large Coach'],axis=1)
target = df[['Supplier Quote(including GST)']]
```

```python
from sklearn.model_selection import cross_validate
from sklearn.linear_model import LinearRegression
# reg = LinearRegression(normalize=True)
# model = cross_validate(reg,features,target,cv=170,return_estimator=True,scoring ='neg_mean_squared_error')
# model
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
x = features.values
y = target.values
```

```python
poly = PolynomialFeatures(degree=3)
```

```python
X_poly = poly.fit_transform(x)
```

```python
poly.fit(X_poly,y)
```

```python
linr = LinearRegression()
```

```python
linr.fit(X_poly,y)
```

```python
preds = linr.predict(X_poly)
preds
```

```python
y
```

```python
from sklearn.metrics import r2_score
r2_score(y,preds)
```

```python
# Calculating percentage difference
```

```python
original=[]
predicted = []
for x in y:
   original.append(x[0])

for x in preds:
    predicted.append(x[0])
```

```python
diff = []
for x in range(len(original)):
    diff.append(predicted[x] - original[x])
    
```

```python
df['Preds'] = predicted
df['Margin Of Error'] = diff
```

```python
df
```

```python
sum = 0
for x in df['Margin Of Error']:
    sum = sum + np.abs(x)

sum/len(df['Margin Of Error'])
```
